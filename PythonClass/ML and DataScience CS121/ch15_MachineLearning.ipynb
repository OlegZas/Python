# Supervised vs Unsupervised ML

## **Supervised Learning**

### **Goal**
Learn a mapping from **input → output** using **labeled data**.

### **What it uses**
- Each training example comes with a **label** (the correct answer).  
  - *Example: An image labeled as “cat.”*

### **What it does**
- Learns patterns that link inputs to known outputs  
- Can then **predict labels** for new, unseen data

### **Common tasks**
#### **1. Classification**
- Predict a **category**  
  - *Example: Spam vs. not spam*

#### **2. Regression**
- Predict a **numeric value**  
  - *Example: Predicting house prices*

### **Examples**
- Predicting whether an email is spam  
- Recognizing handwritten digits (0–9)  
- Predicting a student’s test score from hours studied  

---

## **Unsupervised Learning**

### **Goal**
Find **hidden structure or patterns** in **unlabeled data**.

### **What it uses**
- Training examples **do NOT** come with labels  
- The algorithm must **discover structure** on its own

### **What it does**
- Groups similar items  
- Reduces data complexity  
- Detects anomalies  
- Finds patterns without knowing the “correct answer”

### **Common tasks**
#### **1. Clustering**
- Group similar items  
  - *Example: Group customers by purchasing behavior*

#### **2. Dimensionality Reduction**
- Simplify data while preserving important information  
  - *Example: Compressing images*

#### **3. Association Rules**
- Find relationships  
  - *Example: “People who buy X also buy Y”*

### **Examples**
- Grouping news articles by topic  
- Customer segmentation  
- Detecting unusual credit card transactions  

---

## **Simple Analogy**

### **Supervised Learning → Learning with an answer key**
You study a set of flashcards where each card already has the **correct answer**.  
Later, you’re tested with new cards.

### **Unsupervised Learning → Exploring without instructions**
You get a pile of puzzle pieces with **no picture** on the box.  
You must figure out how the pieces group together on your own.

## **Supervised Machine Learning: Classification (Simplified Explanation)**

**Classification** is a type of supervised machine learning where the goal is to **predict which category (class)** a new data sample belongs to.

### **Binary Classification**
- Involves **two possible classes**  
  - *Example: Is the image a “dog” or a “cat”?*

### **Multi-Class Classification**
- Involves **more than two classes**  
  - *Example: Which digit from 0–9 is shown in an image?*

---

## **Example: Handwritten Digits Dataset**

Scikit-learn includes a dataset of handwritten digits:

- Each sample is an **8×8 pixel image**
- There are **1,797 total images**
- Each image represents a digit from **0 to 9**

### **Goal**
Given one of these images, **predict which digit** it represents.  
This is a **multi-class classification problem** because there are **10 possible classes** (digits 0–9).

---

## **How We Train the Model**

To build a model, we need **labeled data**—images where we already know the correct digit.  
The model learns patterns from these examples so it can classify **new, unseen images**.

---

## **Algorithm Used: k-Nearest Neighbors (k-NN)**

We’ll use **k-nearest neighbors**, one of the simplest classification algorithms.

k-NN works by:

1. Looking at the **k most similar images** (neighbors) in the training data  
2. Letting these neighbors **vote** on which digit the new image represents

This enables the model to **recognize handwritten digits** based on previously seen examples.



## 15.2.1 k-Nearest Neighbors Algorithm (k-NN) 
* Predict a sample’s class by looking at the **_k_ training samples** **nearest in "distance"** to the **sample** 
* Filled dots represent four distinct classes—A (blue), B (green), C (red) and D (purple) 
* **Class with the most “votes” wins**
    * **Odd _k_ value** **avoids ties** &mdash; there’s never an equal number of votes
