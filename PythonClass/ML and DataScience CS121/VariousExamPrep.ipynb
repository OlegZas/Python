```python
import math 


class Point: 
    def __init__(self, x, y):

        self._X = x  
        self._Y = y 

    def distance(self, other):
        return round(math.sqrt((self._X - other._X)**2 + (self._Y - other._Y)**2),2) # rounding bonus

    def slope(self, other):
        return (other._Y - self._Y) / (other._X - self._X)



class Circle:
    def __init__(self, center, edgepoint):
        self._CENTER = center
        self._EDGE = edgepoint
    def area(self):
        return math.pi * (self._CENTER.distance(self._EDGE) ** 2)
    def circumference(self):

        return 2 * math.pi * self._CENTER.distance(self._EDGE) 


class Sphere:
    def __init__(self, circle):
        self._CENTERALCIRCLE = circle
    def volume(self):
        r = self._CENTERALCIRCLE._CENTER.distance(self._CENTERALCIRCLE._EDGE)
        return (4 / 3) * math.pi * r ** 3
    def surfacearea(self):
        r = self._CENTERALCIRCLE._CENTER.distance(self._CENTERALCIRCLE._EDGE)
        return 4 * math.pi * r ** 2

p1 = Point(1,2 ) 
p2 = Point(4,6)

print("Distance:", p1.distance(p2))  
print("Slope:", p1.slope(p2))  
circle = Circle(p1, p2)

print("Area:", circle.area())
print("circumference:", circle.circumference())

sphere = Sphere(circle) 
print("Volume:", sphere.volume()) 
print("Surface Area:", sphere.surfacearea())
```

Using only instances of the  point class, write a class representing a Geometric Sector object. 
You can let the fields be public, but make sure that they don't fall victims to immutability so set the fields to be constant.
Provide an area and a perimeter method (function member).

```python
import math


class Point:
    def __init__(self, X, Y):
        self.X = X
        self.Y = Y

    def distance(self, other):
        return round(math.sqrt((self.X - other.X)**2 + (self.Y - other.Y)**2), 2)

    def slope(self, other):
        return (other.Y - self.Y) / (other.X - self.X)

class Sector:
    def __init__(self, center, edge, angle):
        self.center = center
        self.edge = edge
        self.angle = angle  

    def area(self):
        r = self.center.distance(self.edge)
        return 0.5 * (r ** 2) * self.angle

    def perimeter(self):
        r = self.center.distance(self.edge)
        arc_length = r * self.angle
        return 2 * r + arc_length


p1 = Point(0, 0)
p2 = Point(3, 0)

print("Distance:", p1.distance(p2))
print("Slope:", p1.slope(p2))

sector = Sector(p1, p2, math.pi / 2)

print("Sector Area:", round(sector.area(), 2))
print("Sector Perimeter:", round(sector.perimeter(), 2))
```

Using list comprehension write a function that performs the following mapping

The logic here is that you want to take a list of element and then return another list with each odd indexed element being the partial sum of the input list from the current index until the end divided by where the first number starts in the sequence, the first index being 0.

```python
def map(List):
    n = len(List)
    return [sum(List[i:]) / i for i in range(1, n, 2)]

print(map([1, 2, 3, 4, 5, 6]))
print(map([1, 2, 3, 4, 5, 6,7]))
```
