- average of stock price (AND COUNTS of prices)

```python
import pandas as pd # importing Pandas
import os # module in python Standard library to interact wit OS
os.getcwd() # got an error, so had to check the current directory then change it
os.chdir(r'C:\Users\zasuk\Downloads') # in JP use chdir instead of cd 
# "load up your dataset in it up into memory by opening the file in python with panda given the path on your harddrive"
data = pd.read_csv('oz_stocks.csv') # , names = Date,Open,High,Low,Close,Volume,Dividends,Stock Splits
data = data[['Date', 'Close']] #instructed to only use CLOSE value with every date 
# "What you want to do now, is to use  datetimeLinks to an external site.. What date time will do is take a string representation and convert that into a datetime representation."
# converting to datetime, got a warning that it's modifying the VIEW of data not the original data, will ignore
data['Date'] = pd.to_datetime(data['Date'])
data['Year'] = data['Date'].dt.year # "we only need a YEAR not the whole date;" won't overwrite, will create Year column  
result = pd.DataFrame(data['Year'].unique(), columns=['Year']) # Generate a new data frame with only the unique years as it's values
data
```

```python
result['Average'] = 0.0 # creatig avg column with zeros 
result['Count'] = 0 # tasked to find counts as well. most likely a type but will add just in case 
# loop to iterate and caluclate average 
for index, value in result.iterrows(): # iterrows() method to iterate over row in the result dataframe 
    year = value['Year']  # year column assinged to year var 
    avg = data[data['Year'] == year]['Close'].mean().round(2) # I iwill round here to mkae it look a bit better THIS IS THE PRICE. 
    #len counts number of rows in the dataframe
    count = len(data[data['Year'] == year]) # i am adding count here althoug it's a bit silly because "Now, for every year in your result reference find the average number of stock prices in your data"
    result.loc[result['Year'] == year, ['Average', 'Count']] = [avg, count]#updating columns in the result 
print(result) # rounded since it's a price 
```
