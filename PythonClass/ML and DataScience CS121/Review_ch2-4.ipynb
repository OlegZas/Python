## Calculating Compound Interest

### For each year, the loop displays a formatted string containing the year number and the amount on deposit at the end of that year

```python
from decimal import Decimal
principal = Decimal('1000.00')
rate = Decimal('0.05')

for year in range(1, 11):
    amount = principal * (1 + rate) ** year 
    print(f'{year:>2}{amount:>10.2f}')
    # > means right-align the value. 2 means at least 2 characters wide.
    # So if year is 1, it prints as " 1" (with a space before it), and if it’s 10, it prints as "10". 
    # This keeps the column of years nicely aligned.
```

## Break and Continue statments

```python
#Executing a break statement in a while or for immediately exits that statement.
for number in range(100):
    if number == 10:
        break
    print(number, end=' ')
```

```python
#Executing a continue statement in a while or for loop skips the remainder of the loop’s suite.
for number in range(10):
    if number == 5:
        continue
    print(number, end=' ')
```

# 4.9 Default Parameter Values

You can specify that a parameter has a default parameter value.
When calling the function, if you omit the argument for a parameter with a default parameter value, the default value for that parameter is automatically passed.

```python
def rectangle_area(length=2, width=3):
    """Return a rectangle's area."""
    return length * width
rectangle_area()
```

# Any Number of Arguments 
*, indicating that the function can receive any number of additional arguments.
The * operator, when applied to an iterable argument in a function call, unpacks its elements.


```python
def average(*a):
    return sum(a) / len(a)
average(5, 10)
average(5, 10, 15, 20)
```

```python
grades = [88, 75, 96, 55, 83]
average(*grades) # must use * to unpack the list and get an int 
```

# Modifying global variables

```python
x = 7 # Global variable will not be modified 
def access_global():
    x=3
    print('x printed from access_global:', x)
access_global()
x
```

```python
# to modify global variable, use keyword global 
x = 7 # Global variable will not be modified 
def access_global():
    global x;
    x =3
    print('x printed from access_global:', x)
access_global()
x
```
