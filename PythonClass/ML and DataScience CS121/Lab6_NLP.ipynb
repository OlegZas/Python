```python
import pandas as pd
from textblob import TextBlob
from pathlib import Path

#.1 Loading the dataset 
iphone = pd.read_csv('iphone.csv')  

# checking data 
print(iphone.head()) # need only country and reviewDescription  columns 

# "make 3 subsets of data for each country" 
india = iphone[iphone['country'] == 'India'] # from iphone dataframe getting subset wehre country is india 
usa = iphone[iphone['country'] == 'United States'] # spent a lot of time using 'USA' and wondering why i get 0 polarity. 
japan = iphone[iphone['country'] == 'Japan']


        

# here funciton to calculate the sentiment scores of the reviews 
def get_sentiment(review):
    if isinstance(review, str): # i had an issue with non-string, so will EXCLUDE reviews that are not str 
        blob = TextBlob(review)
        return blob.sentiment.polarity  # check 12.2.5 sentiment polarity: -1 to 1 
    return None 
    
# Initializing variables to hold sentiment values
sentindia = 0
sentusa = 0
sentjapan = 0

# to get average need to get total counts 
countIndia = 0 # len(india) ; problem here is i excluded reviews that were not str and causing error, so have to exclude them from counts. 
countUsa = 0
countJapan = 0

# Looping through each country to get the sentiment 
for review in india['reviewDescription']:
    sentiment = get_sentiment(review)
    if sentiment is not None:  # Above, im returning none if not a string. here excluding it or will get an error. 
        sentindia += sentiment
        countIndia += 1 # since excluded non-str from sentiment, must exclude them from coutns as well 

for review in usa['reviewDescription']:
    sentiment = get_sentiment(review)
    if sentiment is not None:  
        sentusa += sentiment
        countUsa += 1
        
for review in japan['reviewDescription']:
    sentiment = get_sentiment(review)
    if sentiment is not None:  
        sentjapan += sentiment
        countJapan += 1

# Calculating the averages for each. 
avgIndia = sentindia / countIndia 
avgUsa = sentusa / countUsa 
Avgjapan = sentjapan / countJapan 

# Print out the results
print(f"Average Sentiment for India: {avgIndia}")
print(f"Average Sentiment for USA: {avgUsa}")
print(f"Average Sentiment for Japan: {Avgjapan}")

# Drum roll: the most positive country.. is.... : 
maxim = max(avgIndia, avgUsa, Avgjapan)

if maxim == avgIndia:
    print("India is the most positive country based on the reviews.")
elif maxim == avgUsa:
    print("USA is the most positive country based on the reviews.")
else:
    print("Japan is the most positive country based on the reviews.")
```
