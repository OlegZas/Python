## Looping Through List 

```python
reconcile = ['obfuscate','amicable', 'esoteric', 'enigmatic', 'misanthrope']
for reconcile in reconcile:
    print("attempt %s"%(reconcile))
```

```python
reconcile = ['obfuscate','amicable', 'esoteric', 'enigmatic', 'misanthrope']
for reconcile in reconcile:
    print("attempt %s"%(reconcile))
    print(f"here is more {reconcile.title()}, and more \n")
    
print("print")
```

### List first 10 square numbers

```python
# Pyramid 
squares = []
for obfuscate in list(range(1,11)):
    mult = obfuscate ** 2
    squares.append(mult)
    print( squares)
# list of squares 
print("and here is the list:")
squares = []
for obfuscate in list(range(1,11)):
    squares.append(obfuscate ** 2) # each new value is appended to the list of squares 
print( squares)
```

### List Comprehensions: 
- combine code into one line 

```python
obfuscate = [value **2 for value in range(1,11)]
print(obfuscate)
```
