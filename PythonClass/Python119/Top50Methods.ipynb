````python
## üî§ String Methods

### 1. `str.upper()` ‚Äì Convert to uppercase
```python
"hello".upper()  # 'HELLO'
```
### 2. `str.lower()` ‚Äì Convert to lowercase
```python
"HELLO".lower()  # 'hello'
```
### 3. `str.title()` ‚Äì Capitalize each word
```python
"hello world".title()  # 'Hello World'
```
### 4. `str.capitalize()` ‚Äì Capitalize first letter
```python
"python".capitalize()  # 'Python'
```
### 5. `str.strip()` ‚Äì Remove leading/trailing spaces
```python
"  text  ".strip()  # 'text'
```
### 6. `str.lstrip()` ‚Äì Remove leading spaces
```python
"  text".lstrip()  # 'text'
```
### 7. `str.rstrip()` ‚Äì Remove trailing spaces
```python
"text  ".rstrip()  # 'text'
```
### 8. `str.replace(old, new)` ‚Äì Replace substring
```python
"aabb".replace("a", "c")  # 'ccbb'
```
### 9. `str.find(sub)` ‚Äì Find index of substring
```python
"hello".find("e")  # 1
```
### 10. `str.index(sub)` ‚Äì Like find(), but errors if not found
```python
"hello".index("e")  # 1
```
### 11. `str.split()` ‚Äì Split string into list
```python
"a b c".split()  # ['a', 'b', 'c']
```
### 12. `str.join(iterable)` ‚Äì Join iterable into string
```python
"-".join(["a", "b"])  # 'a-b'
```
### 13. `str.startswith(prefix)` ‚Äì Check prefix
```python
"hello".startswith("he")  # True
```
### 14. `str.endswith(suffix)` ‚Äì Check suffix
```python
"hello".endswith("o")  # True
```
### 15. `str.isdigit()` ‚Äì Check if all chars are digits
```python
"123".isdigit()  # True
```
### 16. `str.isalpha()` ‚Äì Check if all chars are letters
```python
"abc".isalpha()  # True
```
### 17. `str.isalnum()` ‚Äì Letters or digits
```python
"abc123".isalnum()  # True
```
### 18. `str.isspace()` ‚Äì Only whitespace?
```python
"   ".isspace()  # True
```
### 19. `str.count(sub)` ‚Äì Count occurrences
```python
"banana".count("a")  # 3
```
### 20. `str.zfill(width)` ‚Äì Pad with zeros
```python
"42".zfill(5)  # '00042'
```

---

## üìã List Methods

### 21. `list.append(x)` ‚Äì Add item to end
```python
lst = [1]
lst.append(2)  # [1, 2]
```
### 22. `list.extend(iterable)` ‚Äì Add multiple items
```python
lst.extend([3, 4])  # [1, 2, 3, 4]
```
### 23. `list.insert(i, x)` ‚Äì Insert at index
```python
lst.insert(1, 99)  # [1, 99, 2, 3, 4]
```
### 24. `list.remove(x)` ‚Äì Remove first occurrence
```python
lst.remove(99)  # [1, 2, 3, 4]
```
### 25. `list.pop([i])` ‚Äì Remove and return item
```python
lst.pop()  # 4, list is now [1, 2, 3]
```
### 26. `list.clear()` ‚Äì Remove all items
```python
lst.clear()  # []
```
### 27. `list.index(x)` ‚Äì Index of x
```python
[10, 20, 30].index(20)  # 1
```
### 28. `list.count(x)` ‚Äì Count x
```python
[1, 1, 2].count(1)  # 2
```
### 29. `list.sort()` ‚Äì In-place sort
```python
nums = [3, 1, 2]
nums.sort()  # [1, 2, 3]
```
### 30. `list.reverse()` ‚Äì In-place reverse
```python
nums.reverse()  # [3, 2, 1]
```
### 31. `list.copy()` ‚Äì Shallow copy
```python
copy = nums.copy()
```

---

## üîê Dictionary Methods

### 32. `dict.get(key)` ‚Äì Safe access
```python
d.get("name")
```
### 33. `dict.keys()` ‚Äì All keys
```python
d.keys()  # dict_keys([...])
```
### 34. `dict.values()` ‚Äì All values
```python
d.values()
```
### 35. `dict.items()` ‚Äì Key-value pairs
```python
d.items()
```
### 36. `dict.update(other)` ‚Äì Update dict
```python
d.update({"age": 30})
```
### 37. `dict.pop(key)` ‚Äì Remove and return by key
```python
d.pop("name")
```
### 38. `dict.popitem()` ‚Äì Remove last pair
```python
d.popitem()
```
### 39. `dict.clear()` ‚Äì Remove all
```python
d.clear()
```
### 40. `dict.setdefault(key, default)` ‚Äì Ensure key
```python
d.setdefault("lang", "Python")
```
### 41. `dict.copy()` ‚Äì Copy dictionary
```python
new_dict = d.copy()
```

---

## üî¢ Set Methods

### 42. `set.add(x)` ‚Äì Add an item
```python
s = {1, 2}
s.add(3)
```
### 43. `set.update(iterable)` ‚Äì Add multiple
```python
s.update([4, 5])
```
### 44. `set.remove(x)` ‚Äì Remove item
```python
s.remove(2)
```
### 45. `set.discard(x)` ‚Äì Remove if exists
```python
s.discard(10)  # No error
```
### 46. `set.pop()` ‚Äì Remove arbitrary
```python
s.pop()
```
### 47. `set.clear()` ‚Äì Remove all
```python
s.clear()
```
### 48. `set.union(other)` ‚Äì Combine sets
```python
s1.union(s2)
```
### 49. `set.intersection(other)` ‚Äì Common values
```python
s1.intersection(s2)
```
### 50. `set.difference(other)` ‚Äì Values in s1 not in s2
```python
s1.difference(s2)
```
````
