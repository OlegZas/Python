# 1. **Pandas:**
   - It's a library(collection of tools) that contains functions and data structures to work with structured data<br>
   - Structured data uses columns and rows, is organized, and has a schema <br>

#### Data structures in Pandas: <br>
       - **Series**: like a single column of data; also stores a column with index 
       - **Dataframe**: holds collections of Series, where each one is treated as a column 
### Functions in Pandas:<br>
    - read_csv() - reads data from csv files and loads them into DataFrame data structure 
    - head() - shows the first 5 rows of the DataFrame by default.
    - describe() - summary statistics like mean, median, sdv, etc.
    - drop() - remove columns or rows from the DataFrame
    - groupby() - similar to SQL, groups data based on columns - often used for aggregations 
    - merge() - used to merge DataFrames 
    - to_csv - saving DataFrame to csv

# 2. NumPy

```python
import numpy as ozzz # will import numpy library to use it's methods 

# method 1 with loadtxt -- this is typically used for NUMERICAL DATA. i can load string as well, but they are badly formatted. 
schedule1 = ozzz.loadtxt(r'C:\Users\zasuk\Desktop\ELAC\Python 119\class Assignments\Half Marathon Golden Training.csv', delimiter =',', dtype=str) 
# here i just called loadtxt assigned to variable schedule .. loadtxt contains arguments such as the path to my csv file.. delimiter that i am using is comma (can be tab, |, or ; )
print(schedule1) # will print here to check the data 

print('*****************************************')
print('************************* The second funciton that can be used is genfromtxt **************************************************8')
print('*****************************************')
#  method 2 with genfromtxt - better suited for mixed data types (int and str)
schedule2 = ozzz.genfromtxt(r'C:\Users\zasuk\Desktop\ELAC\Python 119\class Assignments\Half Marathon Golden Training.csv', delimiter=',', dtype=None, names=True)
# same as above, just genfromtxt function... here i will also add one more argument - names - so numpy will treat first row as names (HEADERS) 
print(schedule2) # again printing to check (PS. i really don't need these meaningless comments.) 
```

# 3. CSV built-in module 

```python
import csv # importing python built-in csv module(won't need to install it) to use it's methods 

# with keyword ensures that the files will be closed once the function runs  
# Open the CSV file for reading using the open() function 
# utf-8-sig is an encoding that is used to avoid issues with excel files that were saved as csv utf-8
with open(r'C:\Users\zasuk\Desktop\ELAC\Python 119\class Assignments\Half Marathon Golden Training.csv', mode='r' , encoding='utf-8-sig') as file: # ALIAS file 
    ozdict = csv.DictReader(file)  # DictReader function is available in csv module and it reads the file into a dictionary>>>  giving object file as argument 
    # the keys of the dictionary = column names 
    # vlaues = rows 
    # It is stored as a dictionary 'ozdict' 
    for oz in ozdict: # now looping through each element of the ozdict to print it 
        print(oz)  # now printing each row as a dictionary with column names as keys
```
