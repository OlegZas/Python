### Slicing a List 
- As with the range() function, Python stops one item before the second index you specify.
- use index [x:x] to slice the list
- without a starting index, python will start with 0 .. example: [:4]- means from 0 to 4th index
- without ending index, it will end with the final exisitng index... example [4:] - from 4 to end

<u>**join():**</u>
- In Python, the join() method is used to join elements of an iterable (like a list or tuple) into a single string, with a specified separator between each element.
- separator.join(iterable)
- Separator: The string that will be placed between the elements. This can be any string, including spaces (" ") or newline characters ("\n").
- Iterable: The collection of elements to join (e.g., a list, tuple, or set). Note that all the elements in the iterable must be strings; if they're not, you'll need to convert them to strings first.

```python
misanthrope = ['0 person that hates people', '1enigmatic - mysterious, difficult to interpret', '2esoteric - only known by a small gorup with specialized interest',
              '4 chastise means карать наказывать']
print(misanthrope[1:3])

print('****************# to print on the new line ')
# to print on the new line 
print('\n'.join(misanthrope[0:2])) # use join

print('**********************************')
# LOOPING THROUGH A SLICED LIST 
for x in misanthrope[2:4] :
    print(x)
```

## Copying a list 
- making a list based on the existing one
- assign exisitng list with a slice [:] to the new list

```python
copiedlist = misanthrope
print('\n'.join(copiedlist))
print('************************************************')
copiedlist.append('adding')
print('\n'.join(copiedlist))
```
