```python
# 1. len()
# Returns the length (number of items) of an object.
len("Python")  # 6

# 2. type()
# Returns the type of an object.
type(42)  # <class 'int'>

# 3. print()
# Outputs data to the console.
print("Hello, world!")  # Hello, world!

# 4. input()
# Reads a line of input from the user.
# name = input("Enter your name: ")  # Uncomment to use in Jupyter

# 5. int()
# Converts a number or string to an integer.
int("10")  # 10

# 6. float()
# Converts a number or string to a floating-point number.
float("3.14")  # 3.14

# 7. str()
# Converts an object to a string.
str(100)  # '100'

# 8. list()
# Creates a list.
list("abc")  # ['a', 'b', 'c']

# 9. dict()
# Creates a dictionary.
dict(a=1, b=2)  # {'a': 1, 'b': 2}

# 10. set()
# Creates a set (unique elements only).
set([1, 2, 2, 3])  # {1, 2, 3}

# 11. sum()
# Returns the sum of all items in an iterable.
sum([1, 2, 3])  # 6

# 12. max()
# Returns the largest item.
max([1, 5, 2])  # 5

# 13. min()
# Returns the smallest item.
min([1, 5, 2])  # 1

# 14. sorted()
# Returns a sorted list from the items in an iterable.
sorted([3, 1, 2])  # [1, 2, 3]

# 15. range()
# Generates a sequence of numbers.
list(range(5))  # [0, 1, 2, 3, 4]

# 16. enumerate()
# Returns an enumerate object (index, item) pairs.
list(enumerate(['a', 'b', 'c']))  # [(0, 'a'), (1, 'b'), (2, 'c')]

# 17. zip()
# Combines two or more iterables into tuples.
list(zip([1, 2], ['a', 'b']))  # [(1, 'a'), (2, 'b')]

# 18. map()
# Applies a function to every item in an iterable.
list(map(str.upper, ['a', 'b']))  # ['A', 'B']

# 19. filter()
# Filters elements using a function that returns True or False.
list(filter(lambda x: x > 0, [-1, 0, 1, 2]))  # [1, 2]

# 20. help()
# Opens the built-in help system.
# help(len)  # Uncomment to see documentation in Jupyter
```
