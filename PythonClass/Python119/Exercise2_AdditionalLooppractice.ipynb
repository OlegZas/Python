**`A loop`** is some process that repeats over and over again. In programming, loops are blocks of codes that will execute while the conditions remain true. Loops allow us to write certain blocks of code, which needs to be repeated, once instead of having 650 ‘If’ statements. 
In python there are three main types of loops and each one of them is important and used regularly. 

1. **While Loop** – they contain conditions and statements that well be executed while conditions are true and what happens once the condition is false and loop is broken. 
2.	**For Loop** – are used for traversing over variables a certain number of times. While the operator is within the specified range - the loop will execute. The loop will stop once the condition becomes false and, if specified, an else statement will run. 
3.	**Nested Loops** – these could be either For Loops or While loops that are inner loops nested inside of outer loops. Some do not consider nested loops to be a separate loop type.


## While Loop (primero)

```python
captious = 22
brazen = 25
wonky = 6

while captious > wonky and captious < brazen :
    print(f"predirchivij {captious}, naglij{brazen}, shatkij{wonky}")
    captious+=1 # in PYthon there is no ++
else :
    print("loopo completo finito")
```

# For Loop (segundo)

for is follwed by the operator that denotes the current element in the list. it will automatically increment with each iteration of the loop. 

```python
ozList = ["lethargic", "incredulous", "precarious"]
for i in ozList: # i is just a current element
    print(i)
else:
    print("Finito!")
```
