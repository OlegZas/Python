```python
guestName = input("Welcome to the Bit & Byte Cafe! \n\nMy name is Oleg and I will be your virtual server today!!! \n\n\t\tWhat shall we call you? - ") # first, i will get the customer's name by using the input function and 
#storing it in the guestname variable 
# print(f"{GuestName}") # just checking to ensure input works correctly 

# To store the menu items i will choose the dictionary where key will be a menu item and price will be the value 
# to follow the 'keep it simple' requirement i will store them all in one list 
Menu = {"1. Full Stack Pasta" : 33.00, "2. Bit Burger": 31.00, "3. Pizza Byte": 30.50, "4. Hello Humus": 19.00, "5. RAM Cola": 6.00, "6. Java Juice": 8.50, "7. Syntax Cake": 14.00, "8. Anaconda Eclair": 16.00}

print(f"\nHola {guestName}!!! Here is our Menu: \n\n") # here will just use print function to add blank lines and the menu title (to improve readibility)
# note: i also tried using \n line breaker, but can't use it inside of the dictionary in python 

# now i will print the menu items for the user. I can just use pint(menu) but that will list all pairs in one line. Instead, i will loop through the dictionary and list them one by one so it's easier to read and choose: 
# i want both price and the name so i will need to use items() method and use two iterators to denote name of the menu item and the price 
for item, price in Menu.items(): # my first for loop that just iterates over my dictionary pairs and prints them one by one. no magic here 
    print(f"{item} : {price}") # cant use + because string can't be concatenated to a float (price). so i will use f string and placeholders - i've done it in every previous exercise. Just returns the output
#print(Menu)

# Now will use print function to prompt user for the input and select the menu items that he/she wants. also a few informaitive messages using the print funciton that prints my outputs 
print("\nPlease select the numbers of the items you'd like to order.")
print("Enter them separated by commas without any spaces (ex: 1,5,8):\n") # i am using \n line breakers a lot to improve readiblity; I know i can format it better, but you want everyone to stick to the absolute basics, so here we are. 

Order = input("\t\tYour choices: ")
orderChoices = Order.split(",") # i need to store the customer's choises so i can calculate the costs later. 
#To do this i will use a method split. All it does is separates a string into elements based on choosen separator and stores as a list
# so my Order object (customer's input) will call the split method to separate the chosen menu items based on commas and store them into a list orderChoices. simple. 
#PS. There are A LOT of formatting and validations that i can do here to minimize the chance of errors but then the professor will deduct points if he thinks using mehtods like strip() or isdigit() is too advanced. will avoid doing so. 

totalCost = 0.00 # now I will declare a variable total_cost and give a default that will be shown if no choices match the menu items 
orderedItems = [] # this is an empty list where i will store the items from the menu that matched the user's order - this is necessary to get the prices and add them up 

# now i need to do multiple things here. First, I need to check if each number in the user's input list (orderChoices) is one of the numbers from the Menu. 
# Match item numbers with menu entries
for number in orderChoices: # first, the for loop to get each item in the user input list 
    found = False # i will add a control variable here (flag). initially will set to false, then in the loop set to true if the number is in the menu. 
    for item in Menu: # second, the for loop to get each item in the menu list 
        if item.startswith(number + "."): # now the conditional if, else statment here to FIND specific items in the menu that match the user input 
            # startswith method just checks if the menu items starts with the user's input number.
            orderedItems.append(item) # if the item is one of the chosen ones, then it will be added to the list i created above specifically for this purpose! 
            # append is a simple mehtod used with list, i used it multiple times in previous assingments, it just adds the element to the list 
            # SUM 
            # now i just need to add an item to the totalCost that is 0 by defulat. but with each iteration it will increase if numbers are added 
            totalCost = totalCost + Menu[item]  # could also use += but will keep it very simple 
            found = True
            break
    if found == False: # originally tried using else within the loop (but that led to incorect output). So here, if the loop did not set found to TRUE, it means the number is not on the list, so it will print 
        # output for the the scenario when numbers are not on the menu 
            print(f"Item {number} is not on the menu.")
            
# Viola! now i just need to print out all the outputs for the customer 
# i will add very basic validation to ensure that there is input and it was stored 
# for this i'll just use conditonal if else statment 
if orderedItems: # TRUE - checks that it's not empty 
    print(f"\nHere's your order summary, {guestName}:") # using print function for the output with the f string to personalize it and print cusotmer name 
    for item in orderedItems: # i will print each ordered item on a separate line for better redibility just like above. for this will iterate over orderedItems with a for loop 
        print(f"- {item}: ${Menu[item]}") # here just printing each specific item that we are iterating and it's price found in the Menu dictionary 
    print(f"\nTotal: ${totalCost}") # this just prints total cost which is calculated above 
else: # FALSE - if orderItems is empty 
    print("\nOh no! You didn't select any valid menu items.")

print("\nThank you for your order! May your latency be low and the bandwidth high!") # just a final goodbye message printed with print function 
```
