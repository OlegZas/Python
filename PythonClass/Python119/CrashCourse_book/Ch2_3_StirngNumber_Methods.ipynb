### <u>**String formatting**</u>
**title()** - capitalize each letter of the string <br>
**upper()** or **lower()** - uppercase or lowercase the string

```python
# capitalize first letter in a string 
equivocate = "verb: use ambigous language as to conceal the truth or avoid commiting oneself."
print(equivocate.title())
example = "Oleg equivocates when he doesn't know the answer."
print(example.upper())

```

### <u>**Using variable value inside a string:**</u>
**f-string**

```python
COMBINE = f"{equivocate} {example}"
print(COMBINE)
```

### <u>**Using New Line and Tab**</u>
**\n** - new line 

**\t** - tab

```python
print(f"{equivocate} \n\t{COMBINE}")
```

### <u>**Removing Whitespaces**</u>

**strip()** - remove white spaces on both sides.<br>
**rstrip()** - remove white spaces on the right<br>
**lstrip()** - remove white spaces on the left<br>

```python
CloudSpanner = " OLTP database, horizontal scalability, globally distributed transactional systems "
SpannerCloud = " ideal for highthroughput, low-latency e-commerce platforms or financial systems " 
print(CloudSpanner.strip())
print(SpannerCloud.lstrip())
```

### <u>**Removing Prefix or Suffix**</u>

**removeprefix('')** - place the prefix in quotation marks (or a variable)

```python
ApacheHadoop = "https: It is an opensource framework that is used to process big data in parallel across a network of computers. \nIt is very good for batch processing. \nIt has high latency, so not good for real-time processing.aaa"
prefix = "https:"
print(f"What is Apache Hadoop?{ApacheHadoop.removeprefix(prefix).removesuffix('aaa')}")
```
