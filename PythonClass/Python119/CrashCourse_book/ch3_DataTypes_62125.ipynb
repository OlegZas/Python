## Formatting data types: 

```python
El_sillon = "sofa or armchair"
print (El_sillon.upper())
print (El_sillon.lower())

armarillo = "closet"
#this is an older way to insert
print("Using operator for %s" % (El_sillon))
#use this way instead: F-string *****************************************
print (f"This is the modern way {El_sillon}") 
#use % for insertion of integers or floats *************** %d and %f
el_gerente = 32
manager = 34434.5343
print("here is an example of integer %d and %f insertion" % (el_gerente, manager))
```

```python
# Using FORMAT function 

toContribute = "aportar"
age = 30
pi = 3.14159

# String
print("toContribute: {}".format(toContribute))    
# Integer
print("Age: {}".format(age))          
# Float with 2 decimal places
print("Pi: {:.2f}".format(pi))

# Multiple values
print("toContribute: {}, Age: {}, Pi: {:.3f}".format(toContribute, age, pi))
# Output: Name: Alex, Age: 30, Pi: 3.142
```

## Type Casting

The term “type casting” refers to the act of converting data from one type to another type.

```python
# int()
int(pi) # this converts the types but DOESN"T STORE IT
pi = int(pi) #now it is STORED 
print(pi)
type(pi)
#float ()
age = float(age)
print(age)
# str()
print(type(str(age)))
print("lets print the above here {} and also {}".format(toContribute, age))
print(f"another way {toContribute} and {pi} done")
print("printing more %s and %f"%(toContribute, pi))
```
