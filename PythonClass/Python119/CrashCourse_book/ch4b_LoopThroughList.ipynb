### <u>3. Make a list of the multiples of 3, from 3 to 30. Use a for loop to print the numbers in your list. </u>

```python
emocionante_exciting = list(range(1, 30))

for i in emocionante_exciting:
    i = i * 3
    if i <= 30:
        print(i)
```

```python
### Alternative to above, one line 

hondo_deep = [orgulloso * 3 for orgulloso in range(1,11)]
print (hondo_deep)
```

```python
## Alternative : 
emocionante_exciting  = [i for i in range(1,31,3)]
print(emocionante_exciting)
```

```python
#Alternative with a regular loop and append method
orgulloso_proud = [] # empty list 
for o in range(3,31,3):
    orgulloso_proud.append(o)
print(orgulloso_proud)
```

```python
# with Lambda 

# Use filter() to select only numbers divisible by 3 from 1 to 30
bello_beautiful = list(filter(lambda x: x % 3 == 0, range(1, 31) ))
# lambda is an anonymous (unnamed) function
# It takes input x and returns True if x is divisible by 3
# This creates a sequence of numbers from 1 to 30
# Print the resulting list of multiples of 3
print(bello_beautiful)
```

### <u>4. List of cubes of 1 to 10 numbers</u>

```python
amistoso_friendly = []
for i in range(1,10):
    amistoso_friendly.append(i**3)
print(amistoso_friendly)

## in line 
amidable = list((i**3 for i in range(1,10)))
print(amidable)

#lambda by itself does not operate over a range automatically
# Have to use map() to apply a lambda function to each item in a list or range:
amidable_friendly = list(map(lambda x: x**3, range(1, 10)))
print(amidable_friendly)
# map() > applies the lambda function to each element in the range

```
