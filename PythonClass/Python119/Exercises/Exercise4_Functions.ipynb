# Functions (exercise 4)

- A function is a block of the reusable code which is only run when you call it. Not all functions have output. 
- It contains a definition and a body which will be executed when the function is called.
- Functions help significantly simplify the code and decrease redundancies by providing a concise way of performing operations. 
### **Functions vs Method**
- The key distinction between a **function** and a **method** is that the former is not associated with the object. **Method** however, is a function that is, and must be, associated with a class.

### Arbitrary parameters 
- used when we don't know the number of arguments that we want to give to a function
- * would use a start to denote arbitrary parameter 

```python
# Arbitrary parameters are used when you don't know how many arguments you'd be passing 

def el_impuesto(*agradable): #*agradable represent xxx number of parameters 
    print(agradable[0]-agradable[1]) # here specifying index of the arguments that will be called 
    print(agradable) # without specifying, will print all 
sano_healthy = (64,343,23,76) # will save tuple as a variable and then give it as an argument to the funciton call  
# el_impuesto(64,343,23,76) # these are the arguments that i provided as a tuple 
el_impuesto(*sano_healthy)# you must use a * before tuple otherwise it will be index out of range; 
# it's because * was used in parameter defention in the el_impuesto function 
```

### Arbitrary keywords 
- keywords can be used within calls when assigning arguments to parameters
- this ensures that the correct values will be assigned to paramter 
- again we don't know how many keywords will be passed into the function 

```python
# here i will use proper naming conventions: 
def keyword_arg(**employeeName):
    print("The employees of the month are:\n" , employeeName['name'],"\n while the worst employee is: ", employeeName['badname'])
    #here using parameter as a placeholder and it will get argumetns based on the keyword match
    # \n starts a new line, using here just to practice and make it look better 
    # i will add a second palceholder with the second keyword to practice 
keyword_arg(name = ['OlegLearning', 'coolOz', 'anakondaOleg'], badname = 'mohamad kudus') #name is the keyword that i set up that will have a list of the best Olegs 
    #interestingly the keyword has to be in ' ' when passed into the placeholder parameter 
    # the second keyword 'badname' just adding here to check how it works 
```
