# Arrays in Python? 

**Arrays** data structures store values of <u>the same data type only </u> and in a specific order (indexed). They are more memory-effecient than lists but are seldom used because of the 'same type' constraint. 
> They are NOT a built-in datatype in Python and require a NUMPY library to import them. 

  1. you would install numpy by running in your cli : pip install numpy
  2. In your ide import it : import numpy as np


```python
import numpy as esoteric #can use aliases! 
# this exercise is better than I first thought. importing numpy! 
wordOftheDay = esoteric.array(["test", "strings"])
print(wordOftheDay)
#very cool. with numpy, you can store a string in an array! Am I going to use arrays? - probably not, but good to know. 
#arrays are indexed so you can access elements (ordered)
print("The second element of the array is: %s" %(wordOftheDay[1]))
```

### Adding elements to an array 

```python
# arrays are also MUTABLE - can be changed 
wordOftheDay[1]= "kadabra"
print(f"changed the second element : {wordOftheDay[1]}") # i am memorizing various ways to use a placeholder f-string vs % (done in previous exercises)
```

### Min and Max of an Array (from numpy)

```python
# you can find minimum and maximum values in an array 
linda = esoteric.array([3,5,75,1])
print (esoteric.min(linda)) # you have to reference the library every time you use it's built-in functions 
```

### Sorting and Array
 **sort** is used for asc sorting <br>
 **flip** is used for desc soritng 

```python
print(esoteric.sort(linda))
linda = esoteric.sort(linda)
print(esoteric.flip(linda)) # NOTE: flip only works on asc sorted arrays!!! otherwise it will not return in desc! 
```

### Mean and Median of array in numpy 

```python
segun = esoteric.array([5,3,6,int(input("test:"))]) # segun - according to ... esoteric - скрытный, тайный 
# testing input here... you can use it! but you must define the data type (int or s)
average = esoteric.mean(segun)
print('here is the average: %.2f'%(average)) # need more practice with floats. using float here (no double in python) .2 denotes decimals  
```

### Dictionary
