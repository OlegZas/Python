<h1><b><u>Type Conversion</h1>


```python
# Modified sample data:
num1 = 10    # int
num2 = 3.14  # float
num3 = 5 + 7j   # complex

# Convert from int to float:
result1 = float(num1)

# Convert from float to int:
result2 = int(num2)

# Convert from int to complex:
result3 = complex(num1)

print(result1)
print(result2)
print(result3)

print(type(result1))
print(type(result2))
print(type(result3))
```

<h3>Python doesn't have a <b>random()</b> function to generate random numbers directly, but it includes a built-in module named random that can be used for generating random numbers.</h3>

```python
import random

print(random.randrange(1, 10))
print(random.randrange(1, 10))
```
